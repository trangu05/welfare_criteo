```python
#Check balance of treatment and control group

raw_criteo['treatment'].value_counts(normalize = True)

#Check balance of features

treatment = raw_criteo[raw_criteo.treatment == 1]
control = raw_criteo[raw_criteo.treatment == 0]

balance = pd.DataFrame({
    'mean_treated': treatment.mean(),
    'mean_control': control.mean(),
    'diff': treatment.mean() - control.mean()
})
balance

```
